# Matplotlib Tutorial

This notebook shows how to create simple and advanced plots with `matplotlib`.

### Imports

We first import `numpy` which helps us create some random data to plot.
Then we import `matplotlib.pyplot` and give it the canonical name `plt`. This module will be used to do the actual plotting.

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Random data for line plot

```python
# Get random y in range [0, 100] for each x in range [0, 100].
rng = np.random.default_rng(1337)
y = rng.integers(low=0, high=100, size=101)
```

### Simple line plot from random data

Given an array of data points `y`, you can plot them by simply calling `plt.plot(y)`.

```python
plt.plot(y)
plt.show()
```

### Adding title, axis labels and legend

Note that the legend has to know the label of the data, which we pass as a kwarg to the plot function.

```python
plt.plot(y, label="y data")
plt.title("Title")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()

plt.show()
```

### Limiting axes and adding grid

Setting axis limits ensures the plot starts and ends at reasonable x- and y-values. A grid can make it easier to read values directly from the plot.
Additionally, we can explicitly pass the x data, which should have the same length as our y data. By default, matplotlib will use the index of each value in y as the x value but often this is not what we want. In this example, we want to plot data with x-values in range $[100, 200]$.

```python
x = np.arange(len(y)) + 100
```

```python
plt.plot(x, y, label="y data")
plt.title("Title")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.xlim(100, 200)
plt.ylim(0, 100)
plt.grid()
plt.legend()

plt.show()
```

### Setting ticks

To manually controll which x- and y-values are shown on the plot axes, simply pass a list of the values to `plt.xticks()` or `plt.yticks()` respectively. Note that the grid will adjust accordingly.

```python
x = np.arange(len(y))
```

```python
plt.plot(x, y, label="y data")
plt.title("Title")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.xlim(0, 100)
plt.ylim(0, 100)
plt.xticks([0, 25, 50, 75, 100])
plt.yticks([0, 25, 50, 75, 100])
plt.grid()
plt.legend()

plt.show()
```

### Multiple lines

We can have multiple lines in a line plot. All we have to do is call `plt.plot()` for each line we want to plot. The figure will remain the same, so all lines appear in a single plot.

```python
y1 = np.arange(101)**2 * 10e-3
y2 = np.arange(101)**2 * 8e-3
y3 = np.arange(101)**2 * 6e-3
y4 = np.arange(101)**2 * 4e-3
```

```python
plt.plot(x, y1, label="y1 data")
plt.plot(x, y2, label="y2 data")
plt.plot(x, y3, label="y3 data")
plt.plot(x, y4, label="y4 data")
plt.title("$x^2 * c$\nfor $c = \{0.01, 0.008, 0.006, 0.004\}$", fontsize="x-large")
plt.xlabel("$x$", fontsize="x-large")
plt.ylabel("$f(x)$", fontsize="x-large")
plt.xlim(0, 100)
plt.ylim(0, 100)
plt.xticks([0, 25, 50, 75, 100])
plt.yticks([0, 25, 50, 75, 100])
plt.grid()
plt.legend(loc="upper center")
plt.show()
```

### Line color and style

Lines in a line plot can be adjusted in many ways, such as changing color, line style or line thickness. Some line styles also place a marker on each data point.

| **Code** | **Description** |
| -- | -- |
| `"blue"` | Draws a blue line. |
| `"b"` | Draws a blue line. |
| `"b-"` | Draws a blue line. |
| `"b--"` | Draws a dashed blue line. |
| `"r."` | Draws red dot markers. |
| `"r,"` | Draws small red dot markers. |
| `"r--."` | Draws a dashed red line with dot markers. |
| `"r--+"` | Draws a dashed red line with plus markers. |

```python
plt.plot(x, y1, "b-", linewidth=2, label="y1 data")
plt.plot(x, y2, "c--", linewidth=2, label="y2 data")
plt.plot(x, y3, "r,", linewidth=1, label="y3 data")
plt.plot(x, y4, "g-.", linewidth=1.5, label="y4 data")
plt.title("$cx^2$\nfor $c = \{0.01, 0.008, 0.006, 0.004\}$", fontsize="x-large")
plt.xlabel("$x$", fontsize="x-large")
plt.ylabel("$f(x)$", fontsize="x-large")
plt.xlim(0, 100)
plt.ylim(0, 100)
plt.xticks([0, 25, 50, 75, 100])
plt.yticks([0, 25, 50, 75, 100])
plt.grid()
plt.legend(loc="upper center")
plt.show()
```
